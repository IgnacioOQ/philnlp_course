
### **Dictionaries**
Like lists, dictionary store multiple elements. Yet, whilst dictionaries store elements in an ordered way, dictionaries store elements as unordered key-value pairs.

- **List Example:** `["Plato", "Aristotle", "Socrates"][427 ,384, 469]`
- **Dictionary Example:** `{"Plato": 427, "Aristotle": 384, "Socrates": 469}`



**Benefit:**

1.   Dictionaries help you label your data. Instead of remembering that index 2 means “Socrates,” you can directly use "Socrates" as a label. Ideal when you need to look things up quickly (like a contact book or glossary)

2.   Dictonaries are much faster to process than lists



````python

# Example: Lists and Dictionaries
philosophers = ["Plato", "Aristotle", "Socrates"]

birth_years = {
    "Plato": 427,
    "Aristotle": 384,
    "Socrates": 469
}

# Print elements from the list and dictionary


print("First philosopher in the list:", philosophers[0])

print( birth_years["Plato"])
````

````python
# Add a new thinker
birth_years["Descartes"] = 1596

# Update existing info
birth_years["Socrates"] = 470  # Slight correction!

# Remove an entry
del birth_years["Descartes"]

# Check if a philosopher is listed
print("Descartes" in birth_years)  # True

````


## **3. Functions and Loops**


Functions and loops help us **reuse** code, and loops allow us to **repeat** actions.




### **Defining Functions**

A function is defined using the keyword  ```def```, followed by:

*   the function name (what you want to call it)
*   a pair of parentheses (possibly with parameters)
*   a colon ```:```  The logic of the function is indented underneath.

Example:
```python
def greet(name):
    return "Hello, " + name + "!"
```

````python
def greet(name):
    return "Hello, " + name + "!"
````

````python

# Function that moves the last two elements of a list to the front
def rotate_list(x):
    return x[-2:] + x[:-2]

# Example list
example_list = [1, 2, 3, 4, 5, 6, 7]

# Test the function
print("Original List:", example_list)
print("Modified List:", rotate_list(example_list))
````

````python
#Example: shift last item to the front, and return this as an output (leaving the input unchanged)
def shiftL(x) :
  return x[-1:] + x[:-1]

#Example: do the same, while overwriting the input
def shiftLx(x) :
  helper = x[-1:]
  x[1:] = x[:-1]
  x[:1] = helper
````

````python
#new variable
k = [1,2,3]
k[:-1]
````

````python
#reset variable
k = [1,2,3]

#Overwriting the input
shiftLx(k)

#show result
k
````

### **Loops**
Loops let us iterate (move through) elements in a structure like a list or dictionary, or repeat an action until a condition is no longer true.

**Types of loops**
* *for loop:* Best when you know what you're looping through (like a list of ideas or names).
* *while loop:* Best when you want to repeat something until a certain condition is no longer true — more like a persistent philosophical inquiry!


````python
# List of philosophers
philosophers = ["Plato", "Aristotle", "Socrates", "Descartes", "Kant"]

# For loop iterating over the list
for philosopher in philosophers:
    print("Philosopher:", philosopher)
````

````python
count = 0

while count < 3:
    print("Still thinking...")
    count += 1
````
